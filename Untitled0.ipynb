# Python Project Data - Supermarket Sales

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Gathering

```python
pyd = pd.read_csv("/content/Python Project Data - Supermarket Sales.csv")
pyd.head(5)
```


# Data Assessing



**Quality issues**

```python
pyd.head(1)
```

```python
pyd.info()
```

Check for duplicate values ​​in a column Invoice ID And delete it




```python
pyd['Invoice ID'].value_counts()
```

```python
pyd.drop_duplicates(subset=['Invoice ID'], keep='first', inplace=True)
```

Verify and modify data types

```python
pyd.info()
```

Change the data type to the appropriate for it in the Date column and the Unit price column


```python
pyd['Date'] = pd.to_datetime(pyd['Date'])
pyd['Unit price'] = pyd['Unit price'].str.extract(r'(\d+\.\d+)').astype(float)
```

Show data description to find outliers

```python
pyd.describe()
```

Handling outliers Outliers

1-Identify and process irrational values ​​in a quantity

Replace negative values ​​with positive values

2-Check for empty values ​​in the unit price column, total column and quantity column.calculating them with the equation

Total=(Quantity*Unit price)+Tax 5%



```python
pyd.loc[pyd['Quantity'] < 0]
```

```python
pyd['Quantity'].replace({-8: 8, -7: 7, -1: 1}, inplace=True)
```

```python
pyd[pyd['Tax 5%' ].isnull()]
```

```python
pyd['Tax 5%'] = pyd['Tax 5%'].fillna(pyd['Total'] - (pyd['Unit price'] * pyd['Quantity']))
```

```python

pyd.loc[pyd['Unit price'].isna(), 'Unit price'] = (
    (pyd['Total'] - pyd['Tax 5%']) / pyd['Quantity']
)

pyd.loc[pyd['Total'].isna(), 'Total'] = (
    (pyd['Unit price'] * pyd['Quantity']) + pyd['Tax 5%']
)
```

Replace incorrect values ​​in the client type column.

We will replace (-) an unregistered agent.

The word "memberr" was replaced with "member"

```python
pyd["Customer type"].unique()
```

```python
pyd["Customer type"] = pyd["Customer type"].replace("-", "unregistered agent")
pyd["Customer type"] = pyd["Customer type"].replace("Memberr", "Member")
```

```python
pyd.info()
```

```python
pyd.describe()
```

Yangon Naypyitaw Mandalay Branch

 so we delete Yangon Naypyitaw Mandalay

# 5- Data Visualization


# Sales Distribution by Branches


A chart showing the sales of each branch to know the best-selling **branch**

```python

plt.figure(figsize=(8, 5))
sns.barplot(x=pyd["Branch"], y=pyd["Total"], estimator=sum, ci=None)
plt.title("Total Sales by Branch")
plt.xlabel("Branch")
plt.ylabel("Total Sales")
plt.show()
```

#Top-Selling Products Chart
To analyze the top-selling products, we will create a bar chart showing the total sales (Total) for each product (Product line).

```python
top_products = pyd.groupby("Product line")["Total"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
sns.barplot(x=top_products.index, y=top_products.values)
plt.xticks(rotation=45)
plt.title("Top Selling Product Lines")
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.show()
```

# Customer Type vs. Rating Analysis
To understand customer satisfaction, we will analyze the average ratings given by different customer types (Customer type). This will help us identify which customers are most satisfied and which need improvement.

```python
customer_ratings = pyd.groupby("Customer type")["Rating"].mean().sort_values(ascending=False)


plt.figure(figsize=(8, 5))
customer_ratings.plot(kind="bar", color=["green", "red"])


plt.xlabel("Customer Type")
plt.ylabel("Average Rating")
plt.title("Customer Satisfaction Analysis (Rating by Customer Type)")
plt.xticks(rotation=0)
plt.ylim(0, 10)
plt.grid(axis="y", linestyle="--", alpha=0.7)


plt.show()
```

#Time Series Analysis: Total Sales Over Time
To analyze how total sales (Total) change over time (Date), we will create a time series plot. This will help us understand seasonal trends and patterns in sales.

```python

pyd["Date"] = pd.to_datetime(pyd["Date"])

daily_sales = pyd.groupby("Date")["Total"].sum()


plt.figure(figsize=(12, 6))
plt.plot(daily_sales, marker="o", linestyle="-", color="b", label="Total Sales")

plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.title("Time Series Analysis: Total Sales Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```
